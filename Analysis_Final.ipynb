```python
import pandas as pd
#Data_Set Path
dataset_path = 'data/processed/Tourism_Weather_Merged.csv'
df = pd.read_csv(dataset_path)

# Analysis 1: Most and Least Visited Destinations
most_visited_dict = df['Destination'].value_counts().to_dict()
most_visited_city = max(most_visited_dict, key=most_visited_dict.get)
most_visited_count = most_visited_dict[most_visited_city]
least_visited_city = min(most_visited_dict, key=most_visited_dict.get)
least_visited_count = most_visited_dict[least_visited_city]

# Print the results for Analysis 1
print("Analysis 1: Most and Least Visited Destinations")
print(f"Most visited Destination: {most_visited_city} (Visited {most_visited_count} times)")
print(f"Least visited Destination: {least_visited_city} (Visited {least_visited_count} times)")

# Analysis 2: Maximum and Minimum Duration of Stays, and Preferred Accommodation
max_duration_row = df.loc[df['Duration (days)'].idxmax()]
min_duration_row = df.loc[df['Duration (days)'].idxmin()]

max_duration = max_duration_row['Duration (days)']
max_duration_location = max_duration_row['Destination']
max_duration_accommodation = max_duration_row['Accommodation type']

min_duration = min_duration_row['Duration (days)']
min_duration_location = min_duration_row['Destination']
min_duration_accommodation = min_duration_row['Accommodation type']

# Print the results for Analysis 2
print("\nAnalysis 2: Maximum and Minimum Duration of Stays, and Accommodation Booked")
print(f"Maximum duration of stay: {max_duration} days in {max_duration_location} with accommodation type: {max_duration_accommodation}")
print(f"Minimum duration of stay: {min_duration} days in {min_duration_location} with accommodation type: {min_duration_accommodation}")

# Analysis 3: Hottest and Coldest Destinations
hottest_destination_row = df.loc[df['Temperature (°C)'].idxmax()]
coldest_destination_row = df.loc[df['Temperature (°C)'].idxmin()]

hottest_destination = hottest_destination_row['Destination']
hottest_month = hottest_destination_row['Month']
hottest_temperature = hottest_destination_row['Temperature (°C)']

coldest_destination = coldest_destination_row['Destination']
coldest_month = coldest_destination_row['Month']
coldest_temperature = coldest_destination_row['Temperature (°C)']

# Create dictionaries for the results
hottest_data = {'Destination': hottest_destination, 'Month': hottest_month, 'Temperature (°C)': hottest_temperature}
coldest_data = {'Destination': coldest_destination, 'Month': coldest_month, 'Temperature (°C)': coldest_temperature}

# Print the results for Analysis 3
print("\nAnalysis 3: Hottest and Coldest Destinations")
print("Hottest Destination:")
print(hottest_data)
print("\nColdest Destination:")
print(coldest_data)

# Analysis 4: Destination with Highest and Lowest Precipitation
max_precipitation_row = df.loc[df['Precipitation (mm)'].idxmax()]
min_precipitation_row = df.loc[df['Precipitation (mm)'].idxmin()]

max_precipitation_destination = max_precipitation_row['Destination']
max_precipitation_month = max_precipitation_row['Month']
max_precipitation_level = max_precipitation_row['Precipitation (mm)']

min_precipitation_destination = min_precipitation_row['Destination']
min_precipitation_month = min_precipitation_row['Month']
min_precipitation_level = min_precipitation_row['Precipitation (mm)']

# Create dictionaries for the results
max_precipitation_data = {'Destination': max_precipitation_destination, 'Month': max_precipitation_month, 'Precipitation (mm)': max_precipitation_level}
min_precipitation_data = {'Destination': min_precipitation_destination, 'Month': min_precipitation_month, 'Precipitation (mm)': min_precipitation_level}

# Print the results for Analysis 4
print("\nAnalysis 4: Destination with Highest and Lowest Precipitation")
print("Destination with Highest Precipitation:")
print(max_precipitation_data)
print("\nDestination with Lowest Precipitation:")
print(min_precipitation_data)

# Analysis 5: Costliest and Least Costly Transportation
max_cost_transportation_row = df.loc[df['Transportation Cost (USD)'].idxmax()]
min_cost_transportation_row = df.loc[df['Transportation Cost (USD)'].idxmin()]

max_cost_transportation_type = max_cost_transportation_row['Transportation type']
max_cost_destination = max_cost_transportation_row['Destination']
max_cost_month = max_cost_transportation_row['Month']
max_cost = max_cost_transportation_row['Transportation Cost (USD)']

min_cost_transportation_type = min_cost_transportation_row['Transportation type']
min_cost_destination = min_cost_transportation_row['Destination']
min_cost_month = min_cost_transportation_row['Month']
min_cost = min_cost_transportation_row['Transportation Cost (USD)']

# Create dictionaries for the results
max_cost_data = {'Transportation Type': max_cost_transportation_type, 'Destination': max_cost_destination, 'Month': max_cost_month, 'Cost (USD)': max_cost}
min_cost_data = {'Transportation Type': min_cost_transportation_type, 'Destination': min_cost_destination, 'Month': min_cost_month, 'Cost (USD)': min_cost}

# Print the results for Analysis 5
print("\nAnalysis 5: Costliest and Least Costly Transportation")
print("Costliest Transportation:")
print(max_cost_data)
print("\nLeast Costly Transportation:")
print(min_cost_data)

# Analysis 6: Descriptive Statistics for Traveler Age
traveler_age = df['Traveler age']

# Measures of Central Tendency
mean_age = traveler_age.mean()
mode_age = traveler_age.mode().iloc[0]  # Mode may return multiple values, so we take the first one
median_age = traveler_age.median()

# Measures of Dispersion
range_age = traveler_age.max() - traveler_age.min()
variance_age = traveler_age.var()
std_dev_age = traveler_age.std()

# Print the results for Analysis 6
print("\nAnalysis 6: Descriptive Statistics for Traveler Age")
print(f"Mean Age: {mean_age:.2f} years")
print(f"Mode Age: {mode_age} years")
print(f"Median Age: {median_age} years")
print(f"Range of Ages: {range_age} years")
print(f"Variance of Ages: {variance_age:.2f}")
print(f"Standard Deviation of Ages: {std_dev_age:.2f}")

# Analysis 7: Most and Least Visited Nationalities
most_visited_nationalities = df['Traveler nationality'].value_counts().idxmax()
least_visited_nationalities = df['Traveler nationality'].value_counts().idxmin()

# Create sets for the results
most_visited_nationalities_set = {most_visited_nationalities}
least_visited_nationalities_set = {least_visited_nationalities}

# Check for ties and add to the sets
most_visited_counts = df['Traveler nationality'].value_counts()
most_visited_ties = most_visited_counts[most_visited_counts == most_visited_counts.iloc[0]].index
least_visited_counts = df['Traveler nationality'].value_counts()
least_visited_ties = least_visited_counts[least_visited_counts == least_visited_counts.iloc[-1]].index

most_visited_nationalities_set.update(most_visited_ties)
least_visited_nationalities_set.update(least_visited_ties)

# Print the results for Analysis 7
print("\nAnalysis 7: Most and Least Visited Nationalities")
print(f"Nationalities with Most Visits to Other Places: {most_visited_nationalities_set}")
print(f"Nationalities with Fewest Visits to Other Places: {least_visited_nationalities_set}")

# Filter the dataframe for the most visited nationalities
most_visited_nationalities_df = df[df['Traveler nationality'].isin(most_visited_nationalities_set)]

# Analysis 8: Most Visited Destinations by Most Visited Nationalities
most_visited_destinations_dict = most_visited_nationalities_df['Destination'].value_counts().to_dict()
most_visited_destination = max(most_visited_destinations_dict, key=most_visited_destinations_dict.get)
most_visited_destination_count = most_visited_destinations_dict[most_visited_destination]

# Get the most visited nationality
most_visited_nationality = most_visited_nationalities_df['Traveler nationality'].iloc[0]

# Print the results for Analysis 8
print("\nAnalysis 8: Most Visited Destinations by Most Visited Nationalities")
print(f"Most visited destination by {most_visited_nationality} nationals: {most_visited_destination} (Visited {most_visited_destination_count} times)")

# Filter the dataframe for the least visited nationalities
least_visited_nationalities_df = df[df['Traveler nationality'].isin(least_visited_nationalities_set)]

# Analysis 9: Least Visited Destinations by Least Visited Nationalities
least_visited_destinations_dict = least_visited_nationalities_df['Destination'].value_counts().to_dict()
least_visited_destinations = dict(sorted(least_visited_destinations_dict.items(), key=lambda item: item[1]))

# Get the least visited nationalities
least_visited_nationalities = list(least_visited_nationalities_set)

# Create a dictionary to store least visited nationalities and their destinations
least_visited_nationalities_destinations = {}
for nationality in least_visited_nationalities:
    least_visited_destination = list(least_visited_destinations.keys())[0]
    least_visited_destination_count = least_visited_destinations[least_visited_destination]
    least_visited_nationalities_destinations[nationality] = {'Destination': least_visited_destination, 'Count': least_visited_destination_count}

# Print the results for Analysis 9
print("\nAnalysis 9: Least Visited Destinations by Least Visited Nationalities")
for nationality, data in least_visited_nationalities_destinations.items():
    print(f"Least visited destination by {nationality} nationals: {data['Destination']} (Visited {data['Count']} times)")

# Analysis 10: Calculate the correlation matrix for the entire dataset
print("\nAnalysis 10: correlation matrix for the entire dataset (Numerical Values)")
correlation_matrix = df.corr(numeric_only=True)
# Print the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)
```
